# Importing Modules

```python
# Import necessary modules
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import plot_model, to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
import joblib
```

# Hyper Parameters

```python
# Model parameters
lr = 0.01
l2_val = 0.001
dropout_val = 0.2

# Dataset parameters
n_samples=1000
n_features=6
noise=0.1
random_state=42

#Training parameters
epochs=1000
batch_size=64
verbose=1
```

# Model Creation

```python
# Directories for base models and visualizations
base_models_dir = "base_models"
visualizations_dir = "model_visualizations"

# Create directories if they don't exist
os.makedirs(base_models_dir, exist_ok=True)
os.makedirs(visualizations_dir, exist_ok=True)
```

```python
# Function to save model and visualization
def save_model_and_plot(model, model_name):
    plot_path = os.path.join(visualizations_dir, f"{model_name}.png")
    model_path = os.path.join(base_models_dir, f"{model_name}.h5")
    
    plot_model(model, to_file=plot_path, show_shapes=True, show_layer_names=True)
    model.save(model_path)
    
    print(f"Network visualization saved: {plot_path}")
    print(f"Model saved: {model_path}")
```

```python
# Initialize the Sequential model
model_1 = Sequential()

# Input layer
model_1.add(Input(shape=(n_features,)))  # Specify input shape
model_1.add(Flatten())  # Flatten input to 1D

# Dynamically add layers with decreasing units
units = 2**n_features  # Starting units
while units > 2:
    # Add a Dense layer with current `units`
    model_1.add(Dense(units=units, activation='relu', kernel_regularizer=l2(l2_val)))
    model_1.add(BatchNormalization())  # Add Batch Normalization
    model_1.add(Dropout(dropout_val))  # Add Dropout
    units //= 2  # Halve the units for the next layer

# Output layer with 2 units
model_1.add(Dense(units=2, activation="sigmoid"))

# Display the model summary
model_1.summary()
```

```python
# Compile the model with an optimizer, loss function, and metrics
model_1.compile(
    optimizer=SGD(learning_rate=lr),  # Stochastic Gradient Descent with learning rate 0.01
    loss=BinaryCrossentropy(),  # Binary cross-entropy loss for binary classification
    metrics=["accuracy"]  # Track accuracy during training and evaluation
)

# Save model_1
save_model_and_plot(model_1, "model_1")
```

```python
# Initialize the Sequential model
model_2 = Sequential()

# Input layer
model_2.add(Input(shape=(n_features,)))  # Specify input shape
model_2.add(Flatten())  # Flatten input to 1D

# Dynamically add layers with 1/4th the previous units
units = 2**n_features  # Starting units
while units > 2:  # Stop when units reach 4
    # Add a Dense layer with current `units`
    model_2.add(Dense(units=units, activation='relu', kernel_regularizer=l2(l2_val)))
    model_2.add(BatchNormalization())  # Add Batch Normalization
    model_2.add(Dropout(dropout_val))  # Add Dropout
    units //= 4  # Divide the units by 4 for the next layer 

# Output layer with 2 units
model_2.add(Dense(units=2, activation="sigmoid"))

# Display the model summary
model_2.summary()
```

```python
# Compile the model with an optimizer, loss function, and metrics
model_2.compile(
    optimizer=SGD(learning_rate=lr),  # Stochastic Gradient Descent with learning rate 0.01
    loss=BinaryCrossentropy(),  # Binary cross-entropy loss for binary classification
    metrics=["accuracy"]  # Track accuracy during training and evaluation
)

# Save model_2
save_model_and_plot(model_2, "model_2")
```

# Dataset Generation

```python
# Function to generate synthetic data
def generate_synthetic_data(n_samples, n_features, noise, random_state):
    X, y = make_regression(
        n_samples=n_samples,
        n_features=n_features,
        noise=noise,
        random_state=random_state
    )
    y_binary = (y > y.mean()).astype(int)  # Convert to binary labels
    return X, y_binary
```

```python
# Function to scale features
def scale_features(X, feature_ranges):
    scaled_features = []
    for i, (name, lower, upper) in enumerate(feature_ranges):
        scaler = MinMaxScaler(feature_range=(lower, upper))
        scaled_feature = scaler.fit_transform(X[:, i].reshape(-1, 1))
        scaled_features.append(scaled_feature.flatten())
    return scaled_features
```

```python
# Function to create sensor metadata
def create_sensor_metadata(feature_ranges, models, thresholds):
    return pd.DataFrame({
        "Sensor Type": [name for name, _, _ in feature_ranges],
        "Model": models,
        "Threshold Value": thresholds,
        "Range": [f"{lower} to {upper}" for _, lower, upper in feature_ranges]
    })
```

```python
# Function to save DataFrame as CSV
def save_dataframe(df, directory, filename):
    os.makedirs(directory, exist_ok=True)
    path = os.path.join(directory, filename)
    df.to_csv(path, index=False)
    print(f"Dataset saved as '{path}'")
```

```python
# Feature details (names, lower, and upper bounds)
features = [
    ("Ultrasonic Sensor", 0, 400),
    ("Flow Sensor", 1, 30),
    ("Humidity Sensor", 0, 100),
    ("Soil Moisture Sensor", 0, 100),
    ("Pressure Sensor", 300, 1100),
    ("Rain Sensor", 0, 1)
]

# Models and thresholds for description
models = ["HC-SR04", "YF-S201", "DHT22", "Capacitive v1.2", "BMP280", "YL-83"]
thresholds = ["40 cm", "2.5 L/min", ">60%", ">90%", ">850 hPa", "0/1"]
```

```python
# Generate data and scale features
X, y_binary = generate_synthetic_data(n_samples, n_features, noise, random_state)
scaled_features = scale_features(X, features)

# Combine scaled features into a DataFrame
df_dataset = pd.DataFrame({features[i][0]: scaled_features[i] for i in range(len(features))})
df_dataset['Output'] = y_binary

# Save dataset
save_dataframe(df_dataset, "datasets", "synthetic_dataset.csv")

# Display the DataFrame
print("Scaled Sensor Data:")
print(df_dataset.head())
```

```python
# Create sensor metadata DataFrame
df_description = create_sensor_metadata(features, models, thresholds)

# Save dataset
save_dataframe(df_description, "datasets", "description.csv")

# Display the DataFrame
print("Sensor Descriptions:")
print(df_description)
```

# Model Training

```python
# Function to load dataset
def load_dataset(path):
    df = pd.read_csv(path)
    X = df.iloc[:, :-1].values  # Features (all columns except the last one)
    y = df['Output'].values     # Target (last column)
    return X, to_categorical(y)  # Binary classification
```

```python
# Function to split and standardize data
def preprocess_data(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # Save the StandardScaler
    scaler_dir = "trained_models"
    scaler_path = os.path.join(scaler_dir, 'scaler.pkl')
    joblib.dump(scaler, scaler_path)
    print(f"Scaler saved successfully at '{scaler_path}'.")
    
    return X_train, X_test, y_train, y_test
```

```python
# Function to train and evaluate model
def train_and_evaluate(model, X_train, y_train, X_test, y_test, epochs, batch_size, verbose, callbacks=None):
    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=epochs,
        batch_size=batch_size,
        verbose=verbose,
        callbacks=callbacks
    )
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    print(f"Test Loss: {loss:.4f}")
    print(f"Test Accuracy: {accuracy:.4f}")
    return history
```

```python
# Function to plot training history
def plot_and_save_history(history, metric, plot_path):
    plt.plot(history.history[metric], label=f'Training {metric.capitalize()}')
    plt.plot(history.history[f'val_{metric}'], label=f'Validation {metric.capitalize()}')
    plt.xlabel('Epochs')
    plt.ylabel(metric.capitalize())
    plt.legend()
    plt.savefig(plot_path, dpi=300, bbox_inches='tight')  # Save with high resolution
    plt.show()
    print(f"Plot saved as '{plot_path}'.")
```

```python
# Function to save trained model
def save_model(model, directory, filename):
    os.makedirs(directory, exist_ok=True)
    path = os.path.join(directory, filename)
    model.save(path)
    print(f"Model saved as '{path}'")
```

```python
# Create directories for saving outputs
plots_dir = "plots"
trained_models_dir = "trained_models"
os.makedirs(plots_dir, exist_ok=True)
os.makedirs(trained_models_dir, exist_ok=True)
```

```python
# Load and preprocess data
X, y = load_dataset(os.path.join('datasets', 'synthetic_dataset.csv'))
X_train, X_test, y_train, y_test = preprocess_data(X, y)
```

```python
# Callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=100,
    restore_best_weights=True,
    verbose=1
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.1,
    patience=5,
    min_lr=1e-6,
    verbose=1
)
callbacks = [
    early_stopping, 
    reduce_lr
]
```

```python
# Train and evaluate Model 1
print("\n--- Training Model 1 ---")
history_1 = train_and_evaluate(
    model_1,
    X_train, y_train, X_test, y_test,
    epochs, 
    batch_size, 
    verbose, 
    callbacks
)
save_model(model_1, trained_models_dir, "trained_model_1.h5")
plot_and_save_history(history_1, 'accuracy', os.path.join(plots_dir, 'accuracy_1_plot.png'))
plot_and_save_history(history_1, 'loss', os.path.join(plots_dir, 'loss_1_plot.png'))
```

```python
# Train and evaluate Model 2
print("\n--- Training Model 2 ---")
history_2 = train_and_evaluate(
    model_2, 
    X_train, y_train, X_test, y_test, 
    epochs, 
    batch_size, 
    verbose, 
    callbacks
)
save_model(model_2, trained_models_dir, "trained_model_2.h5")
plot_and_save_history(history_2, 'accuracy', os.path.join(plots_dir, 'accuracy_2_plot.png'))
plot_and_save_history(history_2, 'loss', os.path.join(plots_dir, 'loss_2_plot.png'))
```

# Model Inferencing

```python
# Function to load and preprocess inference data
def load_and_preprocess_inference_data(dataset_path, scaler=None):
    df = pd.read_csv(dataset_path)
    X_inference = df.iloc[:, :-1].values  # Features (all columns except the last one)
    y_actual = df.iloc[:, -1].values  # Ground truth labels (last column)
    
    # Standardize features using the same scaler used during training
    if scaler:
        X_inference = scaler.transform(X_inference)
    
    return X_inference, y_actual
```

```python
# Function to load trained models
def load_trained_models(model_paths):
    models = {}
    for name, path in model_paths.items():
        models[name] = load_model(path)
        print(f"Model '{name}' loaded from '{path}'")
    return models
```

```python
# Function to perform inference
def perform_inference(models, X_inference):
    predictions = {}
    predictions_classes = {}
    
    for model_name, model in models.items():
        preds = model.predict(X_inference)
        predictions[model_name] = preds
        predictions_classes[model_name] = np.argmax(preds, axis=1)
    return predictions, predictions_classes
```

```python
# Function to display inference results
def display_inference_results(X_inference, y_actual, predictions, predictions_classes):
    print("\n--- Inference Results ---\n")
    for i in range(len(X_inference)):
        print(f"Sample {i+1}:")
        print(f"  Actual Answer: {y_actual[i]}")
        for model_name in predictions_classes:
            print(f"  {model_name} Prediction: {predictions_classes[model_name][i]} (Probabilities: {predictions[model_name][i]})")
        print("-" * 50)
```

```python
# Function to display mismatched results
def display_mismatched_results(y_actual, predictions_classes):
    print("\n--- Mismatched Inference Results ---\n")
    for i in range(len(y_actual)):
        model_mismatches = []
        for model_name, pred_classes in predictions_classes.items():
            if pred_classes[i] != y_actual[i]:
                model_mismatches.append((model_name, pred_classes[i]))
        
        # Display mismatches only
        if model_mismatches:
            print(f"Sample {i+1}:")
            print(f"  Actual Answer: {y_actual[i]}")
            for model_name, prediction in model_mismatches:
                print(f"  {model_name} Prediction: {prediction}")
            print("-" * 50)
```

```python
# Define paths
dataset_path = os.path.join('datasets', 'synthetic_dataset.csv')
trained_model_paths = {
    'Model 1': os.path.join('trained_models', 'trained_model_1.h5'),
    'Model 2': os.path.join('trained_models', 'trained_model_2.h5')
}
scaler_path = os.path.join('trained_models', 'scaler.pkl')  # Path for saved scaler
```

```python
# Load the scaler (ensure the scaler used during training is consistent)
if os.path.exists(scaler_path):
    import joblib
    scaler = joblib.load(scaler_path)
    print("Scaler loaded successfully.")
else:
    print("Warning: Scaler not found. Ensure your data is already standardized!")
    scaler = None
```

```python
# Load and preprocess data
X_inference, y_actual = load_and_preprocess_inference_data(dataset_path, scaler)

# Load trained models
models = load_trained_models(trained_model_paths)
```

```python
# Perform inference
predictions, predictions_classes = perform_inference(models, X_inference)
```

```python
# Display results
display_inference_results(X_inference, y_actual, predictions, predictions_classes)
```

```python
# Display mismatches
display_mismatched_results(y_actual, predictions_classes)
```
